## Introduction

The unemployment rate is one of the most important indicators used to evaluate the health of a country's labour market. In Canada, unemployment rates vary across provinces and demographic groups due to differences in economic structure, industry composition, and population characteristics.

Major economic events, such as recessions or the COVID-19 pandemic, can have significant impacts on employment levels across the country. Understanding these trends helps reveal long-term labour market patterns and regional disparities.

By analyzing historical unemployment data, we can better understand how employment conditions have evolved over time and which population groups may be most affected by economic changes.

## Project Objectives

The objective of this project is to analyze unemployment trends in Canada using historical labour market data from 1976 to the present.

This project aims to:

- Examine long-term unemployment trends at national and provincial levels
- Compare unemployment patterns across Canadian provinces
- Identify demographic differences in unemployment rates over time
- Use data visualizations to clearly communicate key trends and insights

# Data Cleaning

Before performing exploratory data analysis, the dataset is cleaned to ensure consistency, accuracy, and readability.

The cleaning process includes:

1. Renaming columns for clarity  
2. Dropping irrelevant columns  
3. Fixing column text and data types  
4. Handling missing values  
5. Preparing columns for analysis and visualization

### Step 1: Rename Columns

- Column names are simplified for readability and easier referencing in code.
- Spaces and special characters are removed.

```python
df = df.rename(columns={
    "REF_DATE": "year",
    "GEO": "province",
    "Sex": "sex",
    "Age group": "age_group",
    "Employment": "employment",
    "Full-time employment": "full_time",
    "Part-time employment": "part_time",
    "Labour force": "labour_force",
    "Population": "population",
    "Unemployment": "unemployed",
    "Employment rate": "employment_rate",
    "Participation rate": "participation_rate",
    "Unemployment rate": "unemployment_rate"
})

df.columns
```

### Step 2: Clean Text Columns

- Standardize text formatting using `strip()` and `title()`  
- Replace "Both Sexes" with "All" in `sex` column  
- Standardize age group labels for readability in plots  
- Remove extra spaces in `province` names

```python
# Cleaning the columns
df['sex'] = df['sex'].str.strip().str.title()
df['sex'] = df['sex'].replace({'Both Sexes': 'All'})

df['province'] = df['province'].str.strip().str.title()

df['age_group'] = df['age_group'].str.strip().str.title()
df['age_group'] = df['age_group'].replace({
    '15 To 24 Years': '15-24',
    '25 To 54 Years': '25-54',
    '55 To 64 Years': '55-64',
    '15 To 64 Years': '15-64',
    '15 Years And Over': '15+',
    '25 Years And Over': '25+',
    '55 Years And Over': '55+',
    '15 To 19 Years': '15-19',
    '20 To 24 Years': '20-24'
})

print("Sex categories:", df['sex'].unique())
print("Province categories:", df['province'].unique())
print("Age groups:", df['age_group'].unique())
```

### Step 3: Fix Data Types & Convert Rate Columns

- Convert `year` to numeric for plotting trends over time.  
- Convert rate columns (`unemployment_rate`, `employment_rate`) from decimals to percentages for clarity in analysis and visualization.  
- Remove rows with missing `unemployment_rate` to maintain accuracy.

```python
df['year'] = pd.to_datetime(df['year']).dt.year
print(df['year'].head())
```

```python
rate_cols = [
    'unemployment_rate',
    'employment_rate'
]
df[rate_cols] = df[rate_cols]
print(df[rate_cols].head())
```

```python
df = df.dropna(subset=['unemployment_rate'])

print(df['unemployment_rate'].isna().sum())
```

### Step 4: Drop Unused Columns & Final Dataset Check

- Keep only the columns needed for this analysis:  
  `year`, `province`, `sex`, `age_group`, `unemployed`, `labour_force`, `population`, `unemployment_rate`  
- Drop columns that are not relevant (`employment`, `full_time`, `part_time`, `participation_rate`)  
- Verify the dataset is clean and ready for analysis using:
  - `df.head()` for a quick preview  
  - `df.info()` to check rows, columns, and data types  
  - `df.describe()` for summary statistics

```python
df = df[['year', 'province', 'sex', 'age_group', 'unemployed',
         'labour_force', 'population', 'unemployment_rate']]
print(df.head())
```

###Province-Level Averages for Unemployment Rates
*This interactive bar chart allows exploration of average unemployment rates across Canadian provinces.*

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, NumeralTickFormatter
output_notebook()

# Prep
province_rates_plot = df.groupby('province')['unemployment_rate'].mean().reset_index()
province_rates_plot['unemployment_rate'] = province_rates_plot['unemployment_rate'] / 100

source = ColumnDataSource(province_rates_plot)

# Figure
p = figure(x_range=province_rates_plot['province'],
           height=400, width=700,
           title="Average Unemployment Rate by Province",
           x_axis_label='Province', y_axis_label='Unemployment Rate (%)',
           toolbar_location=None, tools="")

# V-Bars
p.vbar(x='province', top='unemployment_rate', width=0.8, source=source, color="skyblue")

# Customizations
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.major_label_orientation = 1.0

p.yaxis.formatter = NumeralTickFormatter(format="0.0%")

show(p)
```

# Unemployment Rate Trends Per Year
*This interactive line chart showing trends over years. Hover shows exact unemployment rate per year.*

```python
from bokeh.models import HoverTool, ColumnDataSource, FixedTicker, NumeralTickFormatter
from bokeh.plotting import figure, show, output_notebook

output_notebook()

# Prep
yearly_trends['unemployment_rate_frac'] = yearly_trends['unemployment_rate'] / 100
yearly_trends_source = ColumnDataSource(yearly_trends)

# Figure
p = figure(height=400, width=800,
           title="Average Unemployment Rate in Canada Over Time",
           x_axis_label='Year', y_axis_label='Unemployment Rate',
           tools="pan,wheel_zoom,box_zoom,reset,save")

# Added lines and points
p.line(x='year', y='unemployment_rate_frac', source=yearly_trends_source, line_width=2, color='skyblue')
p.circle(x='year', y='unemployment_rate_frac', source=yearly_trends_source, size=6, color='blue')


hover = HoverTool(tooltips=[("Year", "@year"), ("Rate", "@unemployment_rate{0.1f}%")])
p.add_tools(hover)

p.yaxis.formatter = NumeralTickFormatter(format="0%")
p.yaxis.ticker = FixedTicker(ticks=[i/100 for i in range(0, 35, 2)])

custom_years = list(range(1975, 2026, 5))
p.xaxis.ticker = FixedTicker(ticks=custom_years)
p.xaxis.major_label_orientation = 1.0

show(p)
```

####Insight #1

1980s: Sharp rise to 12.3% in 1983-1984, likely reflecting the global recession and high interest rates.

####Insight #2

1990s: Peak of 12.7% in 1993, corresponding to the 90s recession. Then spike to 9.5% after the 2008 financial crisis.

####Insight #3

2020: Jump to 11.2% due to the COVID՞19 pandemic.

####Insight #4

These peaks are followed by gradual recoveries, seems to be a cyclical trend

#Unemployment Trends For Each Province Over Time
*Interactive line chart showing unemployment rate trends by Canadian province over time.*

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool, FixedTicker, NumeralTickFormatter
from bokeh.palettes import Category10

output_notebook()

# Percentage format
province_year_rates['unemployment_rate_frac'] = (
    province_year_rates['unemployment_rate'] / 100
)


provinces = province_year_rates['province'].unique().tolist()
colors = Category10[10]

p = figure(
    height=700,
    width=1200,
    title="Unemployment Rate Trends by Province Over Time",
    x_axis_label="Year",
    y_axis_label="Unemployment Rate (%)",
    tools="pan,wheel_zoom,box_zoom,reset,save"
)

for i, province in enumerate(provinces):
    province_data = province_year_rates[
        province_year_rates['province'] == province
    ]

    source = ColumnDataSource(province_data)


    p.line(
        x='year',
        y='unemployment_rate_frac',
        source=source,
        color=colors[i % len(colors)],
        line_width=2,
        legend_label=province
    )


    p.circle(
        x='year',
        y='unemployment_rate_frac',
        source=source,
        color=colors[i % len(colors)],
        size=4
    )

hover = HoverTool(
    tooltips=[
        ("Province", "@province"),
        ("Year", "@year"),
        ("Unemployment Rate", "@unemployment_rate{0.1f}%")
    ]
)
p.add_tools(hover)

# 0%–36% in 2% increments
p.yaxis.ticker = FixedTicker(
    ticks=[i / 100 for i in range(0, 36, 2)]
)
p.yaxis.formatter = NumeralTickFormatter(format="0%")

# Every 5 years
p.xaxis.ticker = FixedTicker(ticks=list(range(1976, 2026, 5)))
p.xaxis.major_label_orientation = 1.0

p.legend.title = "Provinces"
p.legend.location = "top_right"
p.legend.click_policy = "hide"
p.legend.label_text_font_size = "8pt"
p.legend.glyph_height = 10
p.legend.glyph_width = 10
p.legend.spacing = 2
p.legend.padding = 5

show(p)
```

#Population and Labour Force For Each Province
*Interactive pie charts showing each province’s population share of Canada’s labour force through the years*

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Category10
from bokeh.transform import cumsum
import math

output_notebook()

pie_df = province_pop_lf_rates[
    province_pop_lf_rates['year'] == 2023
].copy()


pie_df['share'] = pie_df['labour_force'] / pie_df['labour_force'].sum()
pie_df['angle'] = pie_df['share'] * 2 * math.pi


pie_df['colors'] = [Category10[10][i % len(Category10[10])] for i in range(len(pie_df))]

source = ColumnDataSource(pie_df)


p = figure(
    height=500,
    width=500,
    title="Provincial Share of Canadian Labour Force (2023)",
    tools="hover",
    tooltips=[
        ("Province", "@province"),
        ("Labour Force", "@labour_force{0,0}"),
        ("Share", "@share{0.0%}")
    ]
)

p.wedge(
    x=0, y=0,
    radius=0.4,
    start_angle=cumsum('angle', include_zero=True),
    end_angle=cumsum('angle'),
    line_color="white",
    fill_color='colors',
    source=source
)


p.axis.visible = False
p.grid.grid_line_color = None

show(p)
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Category10
from bokeh.transform import cumsum
import math

output_notebook()

# Filter for one year
pie_df = province_pop_lf_rates[
    province_pop_lf_rates['year'] == 1976
].copy()

# Calculate share and angle using population instead of labour force
pie_df['share'] = pie_df['population'] / pie_df['population'].sum()
pie_df['angle'] = pie_df['share'] * 2 * math.pi

# Assign colors
pie_df['colors'] = [Category10[10][i % len(Category10[10])] for i in range(len(pie_df))]

source = ColumnDataSource(pie_df)

# Create figure
p = figure(
    height=500,
    width=500,
    title="Provincial Share of Canadian Population (1976)",
    tools="hover",
    tooltips=[
        ("Province", "@province"),
        ("Population", "@population{0,0}"),
        ("Share", "@share{0.0%}")
    ]
)

# Draw wedges
p.wedge(
    x=0, y=0,
    radius=0.4,
    start_angle=cumsum('angle', include_zero=True),
    end_angle=cumsum('angle'),
    line_color="white",
    fill_color='colors',
    source=source
)

# Hide axis and grid
p.axis.visible = False
p.grid.grid_line_color = None

show(p)
```

#COVID-19 Unemployment Rate and Recovery Rate
*Interactive bar charts showing provincial unemployment rates in Canada during the COVID‑19 pandemic (2020–2023).*

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, FactorRange, HoverTool
from bokeh.palettes import Category10, Category20

output_notebook()

# Prep
years = covid_impact_all['year'].unique().astype(str).tolist()
provinces = covid_impact_all['province'].unique().tolist()


factors = [(str(year), province) for year in years for province in provinces]


unemployment_values = covid_impact_all.sort_values(['year','province'])['unemployment_rate'].tolist()

if len(provinces) <= 10:
    colors = Category10[len(provinces)]
elif len(provinces) <= 20:
    colors = Category20[len(provinces)]
else:
    colors = [Category20[20][i % 20] for i in range(len(provinces))]

# Province Colours
color_list_for_source = [colors[provinces.index(prov)] for year, prov in factors]


source = ColumnDataSource(data=dict(
    x=factors,
    unemployment=unemployment_values,
    colors_for_bars=color_list_for_source
))

# Figure
p = figure(x_range=FactorRange(*factors), height=400, width=900,
           title="COVID-19 Era Unemployment by Province (2020–2023)",
           y_axis_label="Unemployment Rate (%)",
           toolbar_location=None, tools="")


p.vbar(x='x', top='unemployment', width=0.8, source=source, color='colors_for_bars')


p.xaxis.major_label_orientation = 1.2
p.yaxis.ticker = list(range(0, 36, 2))
p.yaxis.axis_label_standoff = 10


hover = HoverTool(tooltips=[("Year, Province", "@x"), ("Unemployment", "@unemployment{0.1f}%")])
p.add_tools(hover)


show(p)
```

#### Insight #1

The impact was uneven, with provinces dependent on tourism, hospitality, and resource sectors being most affected

#### Insight #2

By 2022՟2023, many provinces returned to prepandemic rates. This demonstrates a fast and strong post-pandemic labor market recovery

#### Insight #3

Regional inequalities are still shown, with Atlantic provinces lagging behind the national average even in 2023

#### Insight #4

Even in 2023, some provinces still struggle with higher unemployment:

• Newfoundland and Labrador: 12.5%

• Prince Edward Island: 8.8%

• New Brunswick: 8.4%



---



## Key Takeaways

### Age Gap in Job Security
Youth (15-24) face higher unemployment, likely due to limited experience and competition for entry-level positions.

### Workforce Planning Considerations
Analysis highlights the need to account for:
- Population growth and demographic shifts  
- Economic trends  
- Regional and age-group disparities  

### Impact of Economic Downturns
Major economic shocks are visible in the data, including:
- 1980s Recession  
- 2008 Financial Crisis  
- 2020 COVID-19 Pandemic  

### Uneven Pandemic Recovery
Recovery speed varied by province:
- Ontario, Quebec, and western provinces recovered fastest  
- Some Atlantic provinces continue to lag behind
